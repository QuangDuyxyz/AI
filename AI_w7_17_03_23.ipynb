```python
#vẽ biểu đồ data
import pandas as pd
url='mtcars.csv'
data=pd.read_csv(url)
data.head()
import matplotlib.pyplot as plt
data.hist()
plt.show()

#mối tương quan giữa các biếns
correlations=data.corr()
print(correlations)
```

```python
import seaborn as sns 
sns.heatmap(correlations, annot=True ) #màu càng đậm giá trị càng thấp 
```

```python
sns.pairplot(data) 
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']), 
          y=data['mpg'])
print('Hệ số intercept là: ', model.intercept_)
print('Hệ số coeffient là: ', model.coef_)

#pt hồi quy 
 # (mpg = -5,344wt+37,285)
```

```python
#độ chính xác 
model.score(X=pd.DataFrame(data['wt']), 
          y=data['mpg']) 
```

```python
# dự đoán 
y_pred = model.predict(X = pd.DataFrame(data['wt']))
print(y_pred)
```

```python
# vẽ đồ thị 
data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9),
          color = 'black', xlim=(0,7))
plt.plot(data['wt'], y_pred, color='yellow')
```

```python
model=linear_model.LinearRegression()
X=pd.DataFrame([data['wt'],data['wt']**2]).T
model.fit(X,y=data['mpg'])
print('Intercept cua mo hinh:',model.intercept_)
print('Coeffient cua mo hinh:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,data['mpg']))
```

```python
import numpy as np
range=np.arange(1.5,5.5,0.1)
x_range=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=x_range)
data.plot(kind='scatter',
          x='wt',
          y='mpg',
          figsize=(9,9),
          color='black',
          xlim=(0,7))
plt.plot(range,y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
model.fit(X=data.loc[:,['wt','hp']], y =data['mpg'])
print('Intercept la:', model.intercept_)
print('coeffient la:', model.coef_)
print('Do chinh xac la:',model.score(data.loc[:,['wt','hp']],data['mpg'])*100,'%')
```
