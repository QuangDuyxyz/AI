```python
import pandas as pd

# create a sample DataFrame
data = {'name': ['John', 'Alice', 'Bob', 'Kate', 'John'],
        'age': [25, 30, 35, 40, 25],
        'gender': ['M', 'F', 'M', 'F', 'M'],
        'country': ['USA', 'Canada', 'USA', 'Australia', 'USA']}
df = pd.DataFrame(data)

# find unique values in the 'name' column
unique_names = df['name'].unique()
print('Unique names:', unique_names)

# find unique values in the entire DataFrame
unique_values = df.nunique()
print('Unique values:\n', unique_values)
```

```python
import pandas as pd

# create a sample DataFrame
data = {'name': ['John', 'Alice', 'Bob', 'Kate'],
        'age': [25, 30, 35, 40]}
df = pd.DataFrame(data)

# define the uppercase function
def uppercase(x):
    return x.upper()

# apply the uppercase function to the 'name' column
df['name_uppercase'] = df['name'].apply(uppercase)

# print the modified DataFrame
print(df)
```
